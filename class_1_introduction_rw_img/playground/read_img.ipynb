```python
import cv2
```

```python
#read image
image = cv2.imread('./colorimg.jpg')
image
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale Image', gray_image)
# Wait until a key is pressed, otherwise, the window will close immediately 
#These two lines are needed for the imshow to work
cv2.waitKey(0)
# Close the image window
cv2.destroyAllWindows()

cv2.imwrite('grayscale_image.png', gray_image)
```

```python
# How to read using matplot where GUI is not there then

import matplotlib.pyplot as plt

#OpenCV has legacy of reading a colorimage as BGR so it has to be converted into rgb
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```
